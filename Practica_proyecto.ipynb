```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_criment = pd.read_csv('Copia de crimenydelito.csv')
```

```python
df_criment.head()
```

```python
df_criment1 = df_criment[['comuna', 'barrio', 'tipo_delito']].groupby(by=['tipo_delito'], as_index = False).agg({'comuna':'count','barrio':'count'})
df_criment1
```

```python
plt.figure(figsize = (40, 10))
sns.barplot(y = 'comuna', x = 'tipo_delito', data = df_criment1);
```
